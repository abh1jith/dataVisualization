### Does something seem out of the ordinary?

We can see that the months Jan, Feb and Dec are the coldest months for all the years; and the months June, July and August seem to be the Hottest months for all the years from 1995 to 2019.
However, we can see that in 2002-June, the temperature seems unusually low comparing to other years in the heatmap. This month could be classified as an "Outlier"

After further research into this outlier case, I found 2 possibilities that might have lead to it:
1: There could be a Data Entry Error in recording the temperature. But, the other sources suggest that this is unlikely.
2: An extreme weather event, such as a heatwave or a cold snap, may have occurred in June 2002, resulting in a temperature outlier. These types of events can cause sudden and significant changes in temperature.

### Plot a histogram of temperature in year 2010 with bin width of 10 degree F 

the "give_me_bins" function we wrote in class might be helpful, but feel free to write your own

```python
def give_me_bins(s, increment): # s is a series and increment is bin width. 

    my_min = s.min()
    my_max = s.max()

    min_for_nice_bins = np.floor(my_min / increment) * increment
    max_for_nice_bins = (np.ceil(my_max / increment) + 1) * increment

    bins = np.arange(min_for_nice_bins, max_for_nice_bins, increment)
    bins
    
    return bins

bins = give_me_bins(data['2010'], 10)
# bins


fig = plt.figure(figsize=(10, 5))

plt.hist(data['2010'], bins=bins)
plt.title('Temperature Histogram - 2010')
plt.xlabel('Temperature (F)')
plt.ylabel('Frequency')
plt.show()
```
